```python
def prettyplot(Y = 0, x = 0,FS = [10,7], Xlabel = 'x',Ylabel = 'y',Title = 'Graph',newFig = True,graphType = "plot",alpha = 1):
    '''Function to plot things, takes Main data to be plotted first, ie y,x where y = f(x)
    '''
    if type(Y) ==int:
        
        return print("Please Add Data")
    if type(x)==int:
        x = range(len(Y))
    if newFig:
        plt.figure(figsize = FS)
    if graphType == "plot":    
        plt.plot(x,Y,alpha = alpha)
    if graphType == "hist":
        plt.hist(Y)
    plt.title(Title)
    plt.xlabel(Xlabel)
    plt.ylabel(Ylabel)
```
